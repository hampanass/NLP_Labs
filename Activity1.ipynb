```python
import re
```

```python
data = """
"UNBELIEVABLY... the driver, Alex, arrivED on 12/05/2024!
He was super helpful. However, the ride cost $45.50 which is pricey.
Contact support@taxi-tech.com or call (555) 019-2834 for refunds.
I was unhappy with the wait time though."
"""
print(data)
```

```python
norm = data.lower()
norm2 = re.sub(r"\.{2,}", "", norm)
print(norm2)
norm3 = re.sub(r"\!","",norm2)
print(norm3)
```

```python
dates = re.findall(r"\b\d{2}/\d{2}/\d{4}\b", norm3)
print(f"Dates Found: {dates}")
```

```python
prices = re.findall(r"\$\d+(?:\.\d+)?", norm3)
print(f"Prices Found: {prices}")
```

```python
emails = re.findall(r"\b[\w\.-]+@[\w\.-]+\.\w+\b", norm3)
print(f"Emails Found: {emails}")
```

```python
phones = re.findall(r"\(\d{3}\)\s\d{3}-\d{4}", norm3)
print(f"Phone Numbers Found: {phones}")
```

```python
agents = re.findall(r"\b\w+er\b", norm3)
print(f"Agents (Doer): {agents}")
```

```python
past_tense = re.findall(r"\b\w+ed\b", norm3)
print(f"Past Tense Actions: {past_tense}")
```

```python
negation_pattern = r"\b(?:un|dis)\w+\b"
negations = re.findall(negation_pattern, norm3)
print(f"Words with negative prefixes: {negations}")
```

```python
phones_bonus = re.findall(r"(?:\(\d{3}\)|\d{3})[-\s]\d{3}-\d{4}", norm3)
print(f"Phone Numbers (Bonus – Both Formats): {phones_bonus}")
```
